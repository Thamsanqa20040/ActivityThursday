```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
foot_ball = pd.read_csv(r'C:\Users\admin\Downloads\dataset - 2020-09-24 (1).csv')
```

```python
print(foot_ball.tail(7))

print(foot_ball.dtypes)
print(foot_ball.shape)
print(foot_ball[['Nationality', 'Club']].head())
```

```python
print(foot_ball.iloc[9])
print(foot_ball.loc[100:110, ['Goals', 'Appearances']])
foot_ball['Goals per Appearance'] = foot_ball['Goals'] / foot_ball['Appearances']
print(foot_ball.head())
```

```python
arsenal_players = foot_ball[foot_ball['Club'] == 'Arsenal']
print(arsenal_players)
high_scorers = foot_ball[foot_ball['Goals'] > 5]
print(high_scorers)
```

```python
top_scorers = foot_ball.nlargest(10, 'Goals')
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_scorers['Name'], top_scorers['Goals'])
plt.xlabel('Player')
plt.ylabel('Goals')
plt.title('Top 10 Players with Highest Number of Goals')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
